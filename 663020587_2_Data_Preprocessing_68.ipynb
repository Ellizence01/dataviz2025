```python
#มีค่าผิดปกติ (outliers)
import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical columns for box plots
numerical_cols = ['age', 'fare', 'sibsp', 'parch']

# Create box plots for each numerical column
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic[numerical_cols])
plt.title('Box plots of numerical variables')
plt.ylabel('Values')
plt.show()
```

* (10 points) จากข้อมูลที่ให้มา เตรียมข้อมูลให้พร้อมโดยการจัดการกับค่าว่างในข้อมูล ให้อธิบายวิธีที่ท่านเลือกใช้ พร้อมยกเหตุผลประกอบ

```python
# Drop rows with missing values in specific columns (e.g., 'age' and 'embarked')
columns_to_consider = ['age', 'embarked']
titanic.dropna(subset=columns_to_consider, inplace=True)

# Display the shape of the DataFrame after dropping missing values
print("Shape of DataFrame after dropping missing values in", columns_to_consider, ":", titanic.shape)
```

* แปลงข้อมูลเชิงคุณภาพให้เป็นตัวเลข \
    i. (10 points) ใช้ one-hot encoding ในการแปลงตัวแปร Embarked \
    ii. (10 points) แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1 \
    iii. (10 points) สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp (number of siblings/spouses aboard) และ Parch (number of parents/children aboard) \
    iv. (10 points) สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น

```python
#i. (10 points) ใช้ one-hot encoding ในการแปลงตัวแปร Embarked
titanic = pd.get_dummies(titanic, columns=['embarked'], prefix='embarked')

# Display the first few rows to see the changes
display(titanic.head())
```

```python
# ii. (10 points) แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1
titanic['sex'] = titanic['sex'].map({'male': 0, 'female': 1})

# Display the first few rows to see the changes
display(titanic.head())
```

```python
# iii. (10 points) สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp (number of siblings/spouses aboard) และ Parch (number of parents/children aboard)
titanic['FamilySize'] = titanic['sibsp'] + titanic['parch']

# Display the first few rows with the new column
display(titanic.head())
```

```python
#iv. (10 points) สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น
titanic['IsAlone'] = (titanic['FamilySize'] == 0).astype(int)

# Display the first few rows with the new column
display(titanic.head())
```

* ทำการ Normalize ข้อมูล \
    i. (10 points) ตัวแปร Fare และ Age ใช้ Min-Max Scaling \
    ii. (10 points) สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1 \
    iii. (10 points) สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)

```python
# Hint ข้อ iii. สามารถใช้ฟังก์ชัน cut ใน pandas ช่วยได้
import pandas as pd
import numpy as np
pd.cut(np.array([1, 7, 5, 4, 6, 3]), 3) # แบ่งข้อมูลทั้งหมดเป็น 3 ช่วง
# [(0.994, 3.0], (5.0, 7.0], (3.0, 5.0], (3.0, 5.0], (5.0, 7.0], (0.994, 3.0]]
# Categories (3, interval[float64, right]): [(0.994, 3.0] < (3.0, 5.0] < (5.0, 7.0]]
pd.cut(np.array([1, 7, 5, 4, 6, 3]), [0,2,5,8]) # แบ่งข้อมูลทั้งหมดตามค่าที่กำหนด
# [(0, 2], (5, 8], (2, 5], (2, 5], (5, 8], (2, 5]]
# Categories (3, interval[int64, right]): [(0, 2] < (2, 5] < (5, 8]]
```

```python
#i. (10 points) ตัวแปร Fare และ Age ใช้ Min-Max Scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
titanic[['fare', 'age']] = scaler.fit_transform(titanic[['fare', 'age']])

# Display the first few rows to see the scaled values
display(titanic.head())
```

```python
#ii. (10 points) สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1
titanic['Fare_Per_Person'] = titanic['fare'] / (titanic['FamilySize'] + 1)

# Display the first few rows with the new column
display(titanic.head())
```

```python
# iii. (10 points) สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)
bins = [0, 12, 18, 50, 100]
labels = ['0-12', '12-18', '18-50', '50-100']
titanic['AgeGroup'] = pd.cut(titanic['age'], bins=bins, labels=labels, right=False)

# Display the first few rows with the new column
display(titanic.head())
```

* (10 points) นำเสนอการแจกแจงหรือทำความเข้าใจตัวแปรสำคัญในข้อมูล เช่น Age, Fare, and Pclass ด้วยการสร้างฮิสโตแกรมหรือแผนภาพกล่อง

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create histograms for Age, Fare, and Pclass
numerical_cols_to_plot = ['age', 'fare', 'pclass']

plt.figure(figsize=(12, 8))
for i, col in enumerate(numerical_cols_to_plot):
    plt.subplot(2, 2, i + 1)
    sns.histplot(data=titanic, x=col, kde=True)
    plt.title(f'Distribution of {col}')

plt.tight_layout()
plt.show()
```

* เมื่อเตรียมข้อมูลพร้อมแล้ว ให้หาความสัมพันธ์ของตัวแปรต่อไปนี้จากข้อมูล ซึ่งอาจนำเสนอด้วยสถิติและแผนภูมิหรือกราฟก็ได้ พร้อมอธิบายสั้น ๆ ว่าพบอะไรหรือสามารถสรุปอะไรได้จากภาพที่ได้ \
    i. (10 points) survived และ pclass \
    ii. (10 points) survived และ sex \
    iii. (10 points) survived และ age \
    iv. (10 points) จำนวนสมาชิกครอบครัวที่รอดชีวิต และ Fare

```python
# i. survived and pclass
print("Survival rate by Pclass:")
print(titanic.groupby('pclass')['survived'].mean())

sns.barplot(x='pclass', y='survived', data=titanic)
plt.title('Survival Rate by Pclass')
plt.ylabel('Survival Rate')
plt.show()

# ii. survived and sex
print("\nSurvival rate by Sex:")
print(titanic.groupby('sex')['survived'].mean())

sns.barplot(x='sex', y='survived', data=titanic)
plt.title('Survival Rate by Sex')
plt.ylabel('Survival Rate')
plt.xticks([0, 1], ['Male', 'Female'])
plt.show()

# iii. survived and age
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic, x='age', hue='survived', kde=True, multiple='stack')
plt.title('Distribution of Age by Survival')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# iv. survived and Fare (considering FamilySize) - Using Fare_Per_Person
plt.figure(figsize=(10, 6))
sns.boxplot(x='survived', y='Fare_Per_Person', data=titanic)
plt.title('Fare Per Person by Survival')
plt.xlabel('Survived')
plt.ylabel('Fare Per Person (Scaled)')
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.show()
```

* (10 points) สร้างตารางสรุปข้อมูลเพื่อหาคุณลักษณะของผู้โดยสารเรือไททานิค และศึกษาเกี่ยวกับการเสียชีวิตของผู้โดยสาร ทำการสรุปผลที่ได้จากตาราง

```python
summary_table = pd.pivot_table(titanic, values='name', index=['pclass', 'sex', 'AgeGroup'],
                               columns='survived', aggfunc='count', fill_value=0)
print("Summary Table of Passenger Characteristics and Survival:")
display(summary_table)


print("\nInterpretation:")
print("This table shows the count of passengers for different combinations of Pclass, Sex, and AgeGroup, categorized by whether they survived (1) or not (0).")
print("By examining the counts, we can see how these factors relate to survival rates.")
```
