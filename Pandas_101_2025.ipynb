````
#Real HW4
# นำเข้าโมดูล drive จาก google.colab เพื่อเข้าถึง Google Drive
from google.colab import drive
# นำเข้าไลบรารี pandas สำหรับจัดการและวิเคราะห์ข้อมูล
import pandas as pd
# นำเข้าโมดูล os สำหรับโต้ตอบกับระบบปฏิบัติการ เช่น การแสดงรายการไฟล์
import os
# นำเข้า defaultdict จาก collections เพื่อช่วยในการสร้าง dictionary ที่มีค่าเริ่มต้น
from collections import defaultdict

# เชื่อมต่อ Google Drive เข้ากับสภาพแวดล้อม Colab
drive.mount('/content/drive')

# กำหนดพาธหลักไปยังโฟลเดอร์ข้อมูลใน Google Drive
base_path = '/content/drive/MyDrive/Python&Data_viz2025/'

# สร้างลิสต์ว่างๆ เพื่อเก็บ DataFrame จากไฟล์ CSV ทั้งหมด
all_data = []

# วนลูปผ่านแต่ละโฟลเดอร์ (แทนแต่ละเดือน) ในพาธข้อมูลหลัก อ่านชื่อไฟล์ทุกตัวที่อยู่ในไฟล์ base_path
for month in os.listdir(base_path):
    # สร้างพาธแบบเต็มไปยังโฟลเดอร์ของเดือนปัจจุบัน
    #/content/drive/MyDrive/Python&DataViz2025/Data/2025-02
    month_path = os.path.join(base_path, month)
#-------------------------------------------------------------------------------
    # ตรวจสอบว่ารายการปัจจุบันเป็นไดเรกทอรี (โฟลเดอร์) หรือไม่
    if not os.path.isdir(month_path):
        # หากไม่ใช่ไดเรกทอรี ให้ข้ามไปยังรายการถัดไป
        continue
    # วนลูปผ่านแต่ละไฟล์ในโฟลเดอร์ของเดือนปัจจุบัน
    for file in os.listdir(month_path):
        # ตรวจสอบว่า 'stop-and-search' อยู่ในชื่อไฟล์หรือไม่
        if 'stop-and-search' not in file:
            # หากไฟล์ไม่ใช่ไฟล์ stop-and-search ให้ข้ามไปยังไฟล์ถัดไป
            continue
        # สร้างพาธแบบเต็มไปยังไฟล์ปัจจุบัน
        filepath = os.path.join(month_path, file)
#-------------------------------------------------------------------------------
        # ใช้บล็อก try-except เพื่อจัดการข้อผิดพลาดที่อาจเกิดขึ้นระหว่างการอ่านไฟล์
        try:
            # อ่านไฟล์ CSV เข้ามาเก็บไว้ใน pandas DataFrame
            df = pd.read_csv(filepath)
            # แยกชื่อเมืองออกจากชื่อไฟล์
            # โดยการแยกชื่อไฟล์ด้วย '-' และนำส่วนที่อยู่หลังปีและเดือน
            # จากนั้นรวมส่วนที่เหลือด้วย '-' และลบช่องว่างที่นำหน้า/ตามหลัง
            file_parts = file.replace('-stop-and-search.csv', '').split('-')
            #['2025', '01', 'bedfordshire']
            #['2025', '01', 'west', 'midlands']
            city = '-'.join(file_parts[2:]).strip()
            # เพิ่มคอลัมน์ใหม่ 'City' ไปยัง DataFrame และเติมชื่อเมืองที่แยกได้ลงไป
            df['City'] = city
            # เพิ่ม DataFrame ที่ประมวลผลแล้วเข้าในลิสต์ all_data
            all_data.append(df)
        # ดักจับข้อยกเว้นใดๆ ที่เกิดขึ้นระหว่างการประมวลผลไฟล์
        except Exception as e:
            # พิมพ์ข้อความแสดงข้อผิดพลาดหากไม่สามารถอ่านไฟล์ได้
            print(f"อ่านไฟล์ไม่ได้: {filepath} : {e}")
#-------------------------------------------------------------------------------
# ต่อ DataFrame ทั้งหมดในลิสต์ all_data เข้าด้วยกันเป็น DataFrame เดียว
# ignore_index=True จะรีเซ็ตดัชนีของ DataFrame ที่รวมเข้าด้วยกัน
combined_df = pd.concat(all_data, ignore_index=True)

# กรอง DataFrame ที่รวมเข้าด้วยกันเพื่อรวมเฉพาะแถวที่ 'Self-defined ethnicity' มีคำว่า 'Asian'
# na=False ทำให้แน่ใจว่าแถวที่มีค่าว่างใน 'Self-defined ethnicity' จะถูกยกเว้น
asian_df = combined_df[combined_df['Self-defined ethnicity'].str.contains('Asian', na=False)]

# นับจำนวนผู้ถูกเรียกค้นที่เป็นคนเอเชียแยกตามเมือง
#{'west-midlands': 2104, 'metropolitan': 2033, 'lancashire': 1301, ..., 'durham': 5}
asian_counts = asian_df['City'].value_counts().to_dict()

# หากมีข้อมูล ให้แสดงผลเมืองที่มีจำนวนน้อยที่สุดและมากที่สุด
if asian_counts:
    # หาเมืองที่มีจำนวนคนเอเชียมากที่สุด
    most_city = max(asian_counts, key=asian_counts.get)
    # หาเมืองที่มีจำนวนคนเอเชียน้อยที่สุด
    fewest_city = min(asian_counts, key=asian_counts.get)
    # พิมพ์เมืองที่มีจำนวนคนเอเชียน้อยที่สุดและจำนวนของพวกเขา
    print(f"เมืองที่มีจำนวนผู้ถูกเรียกค้นที่เป็นคน Asia น้อยที่สุด คือเมือง {fewest_city} จำนวน {asian_counts[fewest_city]} คน")
    # พิมพ์เมืองที่มีจำนวนคนเอเชียมากที่สุดและจำนวนของพวกเขา
    print(f"เมืองที่มีจำนวนผู้ถูกเรียกค้นที่เป็นคน Asia มากที่สุด คือเมือง {most_city} จำนวน {asian_counts[most_city]} คน")
# หาก dictionary asian_counts ว่างเปล่า (ไม่พบคนเอเชีย)
else:
    # พิมพ์ข้อความที่ระบุว่าไม่พบข้อมูลที่เกี่ยวข้อง
    print("ไม่พบข้อมูลที่เกี่ยวข้องกับผู้มีเชื้อสายเอเชีย")

````

````
##My
import pandas as pd
import os

# Create a list of file paths for each month's data
data_paths = [
    '/content/drive/MyDrive/Python&Data_viz2025/2025-01',
    '/content/drive/MyDrive/Python&Data_viz2025/2025-02',
    '/content/drive/MyDrive/Python&Data_viz2025/2025-03',
    '/content/drive/MyDrive/Python&Data_viz2025/2025-04',
    '/content/drive/MyDrive/Python&Data_viz2025/2025-05'
]

# Initialize an empty dictionary to store the counts of Asian individuals for each city
asian_counts = {}

# Loop through each month's data path
for month_path in data_paths:
    try:
        # List all files in the current month's directory
        files = os.listdir(month_path)
        # Filter for "stop-and-search" CSV files
        stop_and_search_files = [f for f in files if 'stop-and-search' in f]

        # Process each stop and search file
        for file in stop_and_search_files:
            # Construct the full file path
            file_path = os.path.join(month_path, file)
            # Read the CSV file into a DataFrame
            df = pd.read_csv(file_path)

            # Extract the city name from the file name
            city_name = file.split('-stop-and-search.csv')[0].split('-')[-1]

            # Filter the DataFrame for Asian individuals
            asian_df = df[df['Self-defined ethnicity'].str.contains('Asian', na=False)]
            # Get the count of Asian individuals
            count = len(asian_df)

            # Update the total count for the city
            if city_name in asian_counts:
                asian_counts[city_name] += count
            else:
                asian_counts[city_name] = count
    except FileNotFoundError:
        print(f"Directory not found: {month_path}")
        continue

# Find the city with the minimum and maximum counts of Asian individuals
if asian_counts:
    min_city = min(asian_counts, key=asian_counts.get)
    min_count = asian_counts[min_city]
    max_city = max(asian_counts, key=asian_counts.get)
    max_count = asian_counts[max_city]
    print(f"The city with the fewest stop and search incidents involving Asian individuals is '{min_city}' with {min_count} incidents.")
    print(f"The city with the most stop and search incidents involving Asian individuals is '{max_city}' with {max_count} incidents.")
else:
    print("No data found to process.")
````

````
# Hw5 (group) แยกทั้ง 5 เดือน เดือนไหนมีจำนวนผู้ถูกเรียกค้นที่เป็นคน Asia น้อยที่สุด และ มากที่สุด
# แก้จาก code เดิม

import pandas as pd
import os

# Create a list of file paths for each month's data
data_paths = [
    '/content/drive/MyDrive/Python&Data_viz2025/2025-01',
    '/content/drive/MyDrive/Python&Data_viz2025/2025-02',
    '/content/drive/MyDrive/Python&Data_viz2025/2025-03',
    '/content/drive/MyDrive/Python&Data_viz2025/2025-04',
    '/content/drive/MyDrive/Python&Data_viz2025/2025-05'
]

# Initialize dictionaries to store total Asian counts for each month
monthly_asian_counts = {}

# Loop through each month's data path
for month_path in data_paths:
    try:
        # Extract the month from the month_path
        month = os.path.basename(month_path)

        # List all files in the current month's directory
        files = os.listdir(month_path)
        # Filter for "stop-and-search" CSV files
        stop_and_search_files = [f for f in files if 'stop-and-search' in f]

        # Initialize count for the current month
        current_month_asian_count = 0

        # Process each stop and search file in the month
        for file in stop_and_search_files:
            # Construct the full file path
            file_path = os.path.join(month_path, file)
            # Read the CSV file into a DataFrame
            df = pd.read_csv(file_path)

            # Filter the DataFrame for Asian individuals
            asian_df = df[df['Self-defined ethnicity'].str.contains('Asian', na=False)]
            # Get the count of Asian individuals in the current file
            count = len(asian_df)

            # Add to the total count for the current month
            current_month_asian_count += count

        # Store the total count for the month
        monthly_asian_counts[month] = current_month_asian_count

    except FileNotFoundError:
        print(f"Directory not found: {month_path}")
        continue

# Find the month with the minimum and maximum total Asian counts
if monthly_asian_counts:
    min_month = min(monthly_asian_counts, key=monthly_asian_counts.get)
    min_count = monthly_asian_counts[min_month]
    max_month = max(monthly_asian_counts, key=monthly_asian_counts.get)
    max_count = monthly_asian_counts[max_month]
    print(f"The month with the fewest stop and search incidents involving Asian individuals is '{min_month}' with {min_count} incidents.")
    print(f"The month with the most stop and search incidents involving Asian individuals is '{max_month}' with {max_count} incidents.")
else:
    print("No data found to process.")
````

````
# prompt: จาก data this_data วนลูปนับว่ามีนิทานที่มีคำว่า แม่ หรือ พ่อ ในชื่อเรื่อง ทั้งหมดกี่เรื่อง เรื่องอะไรบ้าง

count = 0
stories = []
for index, row in this_data.iterrows():
  if 'แม่' in row['title'] or 'พ่อ' in row['title']:
    count += 1
    stories.append(row['title'])

print(f"There are {count} stories with 'แม่' or 'พ่อ' in the title.")
print("The stories are:")
for story in stories:
  print(story)
````

````
# prompt: มีคำดังกล่าวใน นิทานชีวิต หรือ นิทานประจำถิ่นมากกว่ากัน บอกจำนวนด้วย

import pandas
import pandas as pd
import os
from google.colab import drive
from bs4 import BeautifulSoup

# ... (preceding code)

count_life = 0
count_local = 0
for index, row in this_data.iterrows():
  if ('แม่' in row['title'] or 'พ่อ' in row['title']) and row['type'] == 'นิทานชีวิต':
    count_life += 1
  elif ('แม่' in row['title'] or 'พ่อ' in row['title']) and row['type'] == 'นิทานประจำถิ่น':
    count_local += 1

if count_life > count_local:
  print(f"There are more stories with 'แม่' or 'พ่อ' in the title in นิทานชีวิต ({count_life} stories) than in นิทานประจำถิ่น ({count_local} stories).")
elif count_local > count_life:
  print(f"There are more stories with 'แม่' or 'พ่อ' in the title in นิทานประจำถิ่น ({count_local} stories) than in นิทานชีวิต ({count_life} stories).")
else:
  print(f"There are an equal number of stories with 'แม่' or 'พ่อ' in the title in both นิทานชีวิต and นิทานประจำถิ่น ({count_life} stories each).")
````

````
# prompt: คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่อง ว่า แม่ หรือ พ่อ มากที่สุด (พร้อมทั้งอธิบาย code แต่ละบรรทัดด้วย)

# สร้าง dictionary เพื่อเก็บจำนวนนิทานที่มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่องสำหรับแต่ละคนแต่ง
creator_counts = {}
for index, row in this_data.iterrows():
  if 'แม่' in row['title'] or 'พ่อ' in row['title']:
    creator = row['creator']
    if creator not in creator_counts:
      creator_counts[creator] = 0
    creator_counts[creator] += 1

# หาคนแต่งที่มีจำนวนนิทานมากที่สุด
max_count = 0
most_prolific_creator = None
for creator, count in creator_counts.items():
  if count > max_count:
    max_count = count
    most_prolific_creator = creator

print(f"The creator who wrote the most stories with 'แม่' or 'พ่อ' in the title is {most_prolific_creator} with {max_count} stories.")
````
